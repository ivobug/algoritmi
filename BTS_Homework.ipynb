### Primjer 1 - Obilazak binarnog stabla pretraživanja (BST)

**Primjer obilaska binarnog stabla prikazanog na slici.**



```python
# Python program za prikaz obilazaka binarnog stabla
  
# Python klasa koja predstavlja pojedinačni čvor binarnog stabla
# key - ključ ili vrijednost 
class Node: 
    def __init__(self,key): 
        self.left = None
        self.right = None
        self.val = key 
  
  
# Funkcija za in-order obilazak
def printInorder(root): 
  
    if root: 
  
        # rekurzivno idemo na lijevo dijete
        printInorder(root.left) 
  
        # printamo roditelja
        print(root.val), 
  
        # rekurzivno idemo na desno dijete
        printInorder(root.right) 
  
  
  
# Funkcija za post-order obilazak
def printPostorder(root): 
  
    if root: 
  
        # prvo rekurzivno idemo na lijevo dijete
        printPostorder(root.left) 
  
        # rekurzivno idemo na desno dijete
        printPostorder(root.right) 
  
        #  printamo roditelja
        print(root.val), 
  
  
# Funkcija za pre-order obilazak
def printPreorder(root): 
  
    if root: 
  
        # printamo roditelja
        print(root.val), 
  
        # prvo rekurzivno idemo na lijevo dijete
        printPreorder(root.left) 
  
        # Konačno rekrzivno idemo na desno dijete
        printPreorder(root.right) 
       
        
# Driver kod
if __name__ == "__main__":
    root = Node(25) 
    root.left      = Node(15) 
    root.right     = Node(50) 
    
    root.left.left  = Node(10) 
    root.left.right  = Node(22) 
    root.right.left  = Node(35) 
    root.right.right  = Node(70) 
    
    root.left.left.left  = Node(4) 
    root.left.left.right  = Node(12)
    
    root.left.right.left  = Node(18) 
    root.left.right.right  = Node(24)
    
    root.right.left.left  = Node(31) 
    root.right.left.right  = Node(44) 
    
    root.right.right.left  = Node(66) 
    root.right.right.right  = Node(90) 
        
       
    print ("Pre-order obilazak binarnog stabla:")
    printPreorder(root)
  
    print ("\nIn-order obilazak binarnog stabla:")
    printInorder(root)
  
    print ("\nPost-order obilazak binarnog stabla:")
    printPostorder(root)
```

### Primjer 2 - Konverzija binarnog stabla u binarno stablo traženja (BST)

Potrebno je konvertirati binarno stablo u binarno stablo pretraživanja. Konverzija mora biti napravljena na način da se zadrži originalna struktura binarnog stabla.

**Postupak rješavanja:**


1. Kreiranje pomočnog polja arr[] koje čuva inorder obilazak stabla. Potrebno vrijeme je reda O(n).
2. Sortiranje pomočnog polja arr[]. Vrijeme potrebno za sortiranje ovisi o korištenom algoritmu sortiranja. Ovdje se koristi QuickjSort čija složenost je u najgorem slučaju reda O(n^2), a najbolji slučaj je reda O(n log n).
3. Ponovno napravimo inorder obilazak i kopiramo elemente polja u čvorove stabla jedan po jedan. Ovo zahtjeva O(n) vremena.


```python
# Program za konverzijuz BS u BST
# This code is contributed by Nikhil Kumar Singh(nickzuck_007)   
# A binary tree node 
class Node: 
      
    # Constructor to create a new node 
    def __init__(self, data): 
        self.data  = data  
        self.left = None
        self.right = None
        
    
# Helper function to store the inroder traversal of a tree 
def storeInorder(root, inorder): 
      
    # Base Case 
    if root is None: 
        return 
      
    # First store the left subtree 
    storeInorder(root.left, inorder) 
      
    # Copy the root's data 
    inorder.append(root.data) 
  
    # Finally store the right subtree 
    storeInorder(root.right, inorder) 
  
# A helper funtion to count nodes in a binary tree 
def countNodes(root): 
    if root is None: 
        return 0
  
    return countNodes(root.left) + countNodes(root.right) + 1
  
# Helper function that copies contents of sorted array  
# to Binary tree 
def arrayToBST(arr, root): 
    print(arr)
    # Base Case 
    if root is None: 
        return 
    
    # First update the left subtree 
    arrayToBST(arr, root.left) 
  
    # now update root's data delete the value from array 
    root.data = arr[0] 
    arr.pop(0) 
  
    # Finally update the right subtree 
    arrayToBST(arr, root.right) 
  
# Ova funkcija konvertira BS u BSt
def binaryTreeToBST(root): 
      
    # Base Case: Tree is empty 
    if root is None: 
        return 
      
    # Count the number of nodes in Binary Tree so that  
    # we know the size of temporary array to be created 
    n = countNodes(root) 
  
    # Create the temp array and store the inorder traveral  
    # of tree  
    arr = [] 
    storeInorder(root, arr) 
      
    # Sort the array 
    arr.sort() 
  
    # copy array elements back to binary tree 
    arrayToBST(arr, root) 
  
# Inorder ispis
def printInorder(root): 
    if root is None: 
        return
    printInorder(root.left) 
    print (root.data)  
    printInorder(root.right) 

def postOrder(root):
    if root is None: 
        return
    postOrder(root.left)  
    postOrder(root.right) 
    print(root.data) 
 

def preOrder(root):
    if root is None: 
        return
    print (root.data) 
    preOrder(root.left)  
    preOrder(root.right) 
    
    
count_leaf = 0

        
def ispisListova(root):
    global count_leaf
    if root is None:
            return 0
    if((root.left is None) and (root.right is None)):
            print(root.data)
            count_leaf+=1
            return 1
            
    else:
            return(ispisListova(root.left) + ispisListova(root.right))
                
##  rekurzivni minimum i maximum
def minValue(root):
    if root.left == None:
        return root.data
    return minValue(root.left) 
        
def maxValue(root):
    if root.right == None:
        return root.data
    return minValue(root.right) 
        
def get_tree_height(node):
    if node is None: 
        return 0 
    else : 
        # Compute the height of each subtree  
        lheight = height(node.left) 
        rheight = height(node.right) 
  
        #Use the larger one 
        if lheight > rheight : 
            return lheight+1
        else: 
            return rheight+1
        
def ObilazakStablaQueue(root):
    # Base Case 
    if root is None: 
        return
      
    # Create an empty queue for level order traversal 
    queue = [] 
  
    # Enqueue Root and initialize height 
    queue.append(root) 
    return(queue)


# Function to  print level order traversal of tree 
def printLevelOrder(root): 
    h = height(root) 
    for i in range(1, h+1): 
        printGivenLevel(root, i) 
  
  
# Print nodes at a given level 
def printGivenLevel(root , level): 
    if root is None: 
        return
    if level == 1: 
        print ("%d" %(root.data)), 
    elif level > 1 : 
        printGivenLevel(root.left , level-1) 
        printGivenLevel(root.right , level-1) 
  
""" Compute the height of a tree--the number of nodes 
    along the longest path from the root node down to 
    the farthest leaf node 
"""
def height(node): 
    if node is None: 
        return 0 
    else : 
        # Compute the height of each subtree  
        lheight = height(node.left) 
        rheight = height(node.right) 
  
        #Use the larger one 
        if lheight > rheight : 
            return lheight+1
        else: 
            return rheight+1

        
# Iterative Method to print the height of binary tree 
def printLevelOrder1(root): 
    # Base Case 
    if root is None: 
        return
      
    # Create an empty queue for level order traversal 
    queue = [] 
  
    # Enqueue Root and initialize height 
    queue.append(root) 
  
    while(len(queue) > 0): 
        # Print front of queue and remove it from queue 
        print (queue[0].data), 
        node = queue.pop(0) 
  
        #Enqueue left child 
        if node.left is not None: 
            queue.append(node.left) 
  
        # Enqueue right child 
        if node.right is not None: 
            queue.append(node.right) 
             
        
        
        
# Driver program za testiranje funkcije
if __name__ == "__main__":
    
    """
    Ulaz binarno stablo:
       
          10
         /  \        a)
        2    7
       / \
      8   4
       
Izlaz BST:

          8
         /  \
        4    10      b)
       / \
      2   7
    
    
     
    """
    # zadavanje BINARNOG STABLA prema slici a) - nakon pretvorbe u BST dobijemo sliku b)
    root = Node(10) 
    root.left = Node(2) 
    root.right = Node(7) 
    root.left.left = Node(8) 
    root.right.right= Node(4) 

    print("Inorder ispis binarnog stabla")
    printInorder(root) 
    
    print ("Level order traversal of binary tree is -")
    printLevelOrder(root) 
  
    print ("Queu - Level order traversal of binary tree is -")
    printLevelOrder1(root) 
    
    print("Poziv funkcije za konverziju:")
    # Konverzija BS u BST
    binaryTreeToBST(root) 
    
    print ("PreOrder obilazak konvertiranog BST")
    preOrder(root)
    print ("Inorder obilazak konvertiranog BST")
    print ("Vidimo da je Inorder obilazak BST sortirani niz za razliku od obilaska BS ")
    printInorder(root) 
    print ("PostOrder obilazak konvertiranog BST")
    postOrder(root) 
  

# VLASTITI TESTOVI NAKON IMPLEMENTACIJE ZADANIH FUNKCIJA
    print("Vlastiti testovi \n")
    print("ISPIS LISTOVA")
    ispisListova(root)
    print("BROJ LISTOVA:", count_leaf)

    minim = minValue(root)
    print("minimum ")
    print(minim)
    maks = maxValue(root)
    print("maksimum ")
    print(maks)

    print("visina")
    print( get_tree_height(root))
    
```
